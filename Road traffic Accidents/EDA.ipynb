```python
conn = sqlite3.connect('C:/Users/chiso/Downloads/accident_data_v1.0.0_2023.db')

queries = ["SELECT * FROM accident WHERE accident_year = 2020 LIMIT 5", "SELECT * FROM vehicle WHERE accident_year = 2020 LIMIT 5", "SELECT * FROM casualty WHERE accident_year = 2020 LIMIT 5", "SELECT * FROM lsoa LIMIT 5" ]

acc_data = pd.read_sql_query(queries[0], conn)
veh_data = pd.read_sql_query(queries[1], conn)
cas_data = pd.read_sql_query(queries[2], conn)
lsoa_data = pd.read_sql_query(queries[3], conn)
# Display the DataFrame
print("accident data", acc_data, "\n\n")
print("vehicle data",veh_data, "\n\n")
print("casualities data",cas_data, "\n\n")
print("lsoa data",lsoa_data, "\n\n")
    
# Close the connection
conn.close()
```
